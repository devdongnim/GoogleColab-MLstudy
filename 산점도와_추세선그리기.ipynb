```python
import numpy as np
import matplotlib.pylab as pylab

strain = [0, 180, 1562, 2790]
stress = [0, 7.84, 94.08, 156.50]

# 추세선을 위한 계산 - 1차원의 polynomial(다항식)을 계산하기 위한 코드입니다.
z = np.polyfit(strain, stress, 1) # (X,Y,차원) 정의
p = np.poly1d(z) # 1차원 다항식에 대한 연산을 캡슐화

# 그래프 그리기
pylab.plot(strain,stress,'o') #산점도를 뜻할 때 'o'라고 합니다.
pylab.plot(strain,p(strain),"r--")
pylab.xlabel("strain [10^-6 ϵ]", labelpad=10)
pylab.ylabel("stress [10^6 Pa]", labelpad=10)
```
